```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
dat = cv2.imread('500_3000.png', cv2.IMREAD_GRAYSCALE)
print(dat)
```

```python
img = np.where(dat>256/2.2+20,1,0)

print(img)
```

```python
plt.imshow(img,cmap='gray')
plt.imsave('mask.png',img,cmap='gray')
```
